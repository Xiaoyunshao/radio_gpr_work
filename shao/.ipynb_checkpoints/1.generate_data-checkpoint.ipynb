# load the meta-data

The data is of shape (Nbaseline, Ntimes, Nfrequencies, Npolarizations) -> (Nbls, Ntimes, Nfreq, Npol) -> (243, 1600, 600, 4).

The data is complex.

Each "baseline" is formed between two antennas, so a baseline index is often denoted as (ant1, ant2).

Simulation frequencies are in Hz, times are in Julian Date (which is just in days). LST is local sidereal time in radians(天空转到了哪个方向，同一片天空,今天LST = 8h,明天LST = 8h，虽然 JD 不同，但看到的银河、点源、EoR 区域几乎相同。). Polarization is the polarization state of the electric field, which comes in dual linear polarization state (XX, YY, XY, YX). We also call this (EE, NN, EN, NE) for east and north.

The simulation contains two things:
1. a simulation of the "foregrounds" including the synchrotron emission from the Milky Way galactic disk, and the radio galaxy point source background.

2. a simulation of a mock EoR 21cm field. 

so y = fg + eor for "sum".

The "sum.uvh5" has the two summed together. The "eor.uvh5" has just the EoR simulation. We will generate the noise ourselves in the notebook (using this formalism `hera_cal.noise.predict_noise_variance_from_autos`). 


The simulation has 1600 time integrations and 600 frequency bins, and 243 unique baseline vectors.

We call the data the "visibility" which we also sometimes call the "cross-correlatoin visibility" i.e. the cross-correlation between two antennas. The visibility of an antenna with itself (e.g. 0, 0) we call the "auto-correlation" visibility.

We observe at MHz frequencies (120 - 190 MHz).

The data are in units of Jansky (Jy). -> 1 Jy=10^(−26)Wm^(−2)Hz^(−1)

## 0. generate noise

```python
#https://github.com/HERA-Team/hera_cal/blob/main/hera_cal/noise.py
#def predict_noise_variance_from_autos
#(0,0)=(1,1)
#simulation 只有0，0；无(1,1);(2,2)
#假设相等
#if unit correct for variance,dirrectly add?

import numpy as np
import torch
from hera_cal.noise import predict_noise_variance_from_autos
from hera_cal.datacontainer import DataContainer

# ----------------------------
# 1. 选择你要处理的 baseline / pol
# ----------------------------
  
auto_ant1 = 0
auto_ant2 = 1

bl = (auto_ant1, auto_ant2, 'ee')  #(0,1, 'ee')
#->sigma^2 =V_{01}^2/Delta_t * Delta_v

auto_ant=0
#-> sigma^2 =V_{00}^2/Delta_t * Delta_v

# ----------------------------
# 2. 读取两个 auto-correlation visibility
# ----------------------------
auto00, meta00 = gp_utils.load_uvdata(
    '/Volumes/YUNYUNShao/gpr_data/zen.h6c_idr2_validation.sum.uvh5',
    bls=[(auto_ant1, auto_ant1)],
    time_ints=time_ints,
    freq_chans=freq_chans,
    pols=[bl[2]],  # 'ee'
    device=device
)





# auto shape 通常是:
# (1, Ntimes, Nfreq, Npol)

# ----------------------------
# 3. auto00[0, :, :, 0] -> [ :, :]
# ----------------------------
auto00_ee = auto00[0, :, :, 0].cpu().numpy()

# ----------------------------
# 4. 构造 DataContainer
# ----------------------------
auto_data = DataContainer({
    (auto_ant1,auto_ant1, bl[2]): auto00_ee, #(0, 0, 'ee'): auto00_ee,
})


freqs00 = meta00['freqs']/1e6  #HZ -> MHZ
times00 = meta00['times']  #value: day

Ntimes00 = len(times00)


# 手动给 dt, df 最稳
# df: channel width, Hz
# dt: integration time, seconds

df = np.median(np.diff(freqs00)) * 1e6              # freqs 单位 MHZ -> Hz
dt = np.median(np.diff(times00)) * 24 * 3600   # 如果 times 是 JD, 单位 day -> second 


# ----------------------------
# 5. 预测 baseline (3,5,'ee') 的 noise variance
# ----------------------------
noise_var = predict_noise_variance_from_autos(
    bl,
    auto_data,
    dt=dt,
    df=df,
    auto_ant=auto_ant,   #auto_ant=0  
)

print(noise_var.shape)   # (Ntimes, Nfreq)

# ----------------------------
# 6. 生成 complex Gaussian noise
# ----------------------------
noise_var_torch = torch.as_tensor(noise_var, device=device, dtype=torch.double)

# 变成和 data 一样的 shape: (Nbls, Ntimes, Nfreq, Npol)
noise_var_torch = noise_var_torch[None, :, :, None]

noise = (
    torch.randn_like(noise_var_torch)
    + 1j * torch.randn_like(noise_var_torch)
) * torch.sqrt(noise_var_torch / 2)


print(noise.shape)
```

```python
# ----------------------------
# 7. save
# ----------------------------
savefile = "shao/generate_data/noise.npz"

np.savez(
    savefile,
    noise=noise.cpu().numpy(),
    noise_var=noise_var_torch.cpu().numpy(),
    freqs=freqs00,
    times=times00,
)

print(f"saved to {savefile}")
```
